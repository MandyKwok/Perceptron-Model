```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
```

```python
#生成数据 (2,1000) (1,1000)
X,y = make_classification(n_samples=1000, n_features=2,n_redundant=0,n_informative=1,n_clusters_per_class=1)
X_train,X_test,y_trian,y_test = train_test_split(X,y,test_size = 0.2,random_state = 33)
```

```python
#使用默认参数拟合模型
p_clf = Perceptron()
p_clf.fit(X_train,y_trian)
print(p_clf.coef_,p_clf.intercept_)
```

```python
print('准确率 {}'.format(p_clf.score(X_test,y_test)))
```

```python
#绘图显示
#将数据进行分类 f1,f2为样本的特征
positive_f1 = [X[i,0] for i in range(1000) if y[i] == 0]
positive_f2 = [X[i,1] for i in range(1000) if y[i] == 0]
negetive_f1 = [X[i,0] for i in range(1000) if y[i] == 1]
negetive_f2 = [X[i,1] for i in range(1000) if y[i] == 1]

plt.figure()
plt.scatter(positive_f1,positive_f2,c = 'red',alpha=0.5)
plt.scatter(negetive_f1,negetive_f2,c = 'blue',alpha=0.5)
"""
w1*f1 + w2*f2 + b = 0
f2 = (-b - w1*f1)/w2
f2 = -f1*(w1/w2) -b/w2
"""  
line_f1 = np.arange(-0.1,1.5)
line_f2 = -line_f1*(p_clf.coef_[0][0]/p_clf.coef_[0][1])-(p_clf.intercept_/p_clf.coef_[0][1])
plt.plot(line_f1,line_f2)
plt.show()
```
